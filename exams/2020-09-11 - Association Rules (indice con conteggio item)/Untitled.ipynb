```python
# imports and setup
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.evaluate import accuracy_score
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules

%matplotlib inline

random_state=777

np.random.seed(random_state)
```

### 0. read a market basket database from the csv file provided and generate a dataframe basket of boolean values with one row per transaction and one column per distinct item of the atabase; the dataframe values must be True if a distinct item is contained in the transaction

- the file contains one transaction per line, the first
element is the number of items in the transaction,
followed by the items of the transaction, and then a
variable number of empty fields
- the field names in the first row of the csv file are not
relevant

```python
f = open('./mba.csv', mode='r')

transactions = [line.strip('\n').split(',') for line in f.readlines()]

transactions = transactions[1:]

for i in range(len(transactions)):
    # The end index is the number of items in the transaction
    # plus one, because we have the number itself in the slot 0
    end_index = int(transactions[i][0]) + 1
    transactions[i] = transactions[i][1:end_index]

encoder = TransactionEncoder()

encoded = encoder.fit_transform(transactions)

basket = pd.DataFrame(encoded)

print(f"len of basket {basket.shape[0]}")

basket.sample(10, random_state=random_state)
```

### 1. ignore the transactions containing a single item (2 points)

```python
single_item_transactions_idxs = []

for i, trans in enumerate(transactions):
    if len(trans) == 1:
        single_item_transactions_idxs.append(i)

basket.drop(index=single_item_transactions_idxs, inplace=True)

print(f"len of basket {basket.shape[0]}")

basket
```

### 2. the column names of the output dataframe are the distinct items (2 points)

```python
columns = encoder.columns_

basket.columns = columns
```

### 3. show the first five rows of the output dataframe (1 point)

```python
basket.head(5)
```

### 4. show the number of transactions and of distinct items (1 point)

```python
print(f"total transaction number: {basket.shape[0]}")
print(f"total distinct items number: {len(columns)}")
```

### 5. find a value of min_support such that the apriori algorithm generates at least 8 frequent itemsets with at least 2 items (5 points)
- output the result with the message below
- min_support: 0.xxxx - number of itemsets with at least 2 items: nn)

```python
# requirements
min_frequent_itemsets_count_threshold = 8
min_items_in_itemset_threshold = 2

min_support = .2

step = 0.01

while min_support > 0:
    frequent_itemsets = apriori(basket, min_support=min_support, use_colnames=True)
    
    itemsets_above_threshold = sum([len(itemset) >= min_items_in_itemset_threshold for itemset in frequent_itemsets.itemsets])

    if itemsets_above_threshold >= min_frequent_itemsets_count_threshold:
        break
    
    min_support -= step

if min_support == 0:
    print(f"nothing found that matches the requirements, try with a bigger range")
else:
    print(f"found min_support for the given requirements")
    print(f"min_support={min_support:.4f} - number of itemsets with at least 2 items: {itemsets_above_threshold}")
```

### 6. find the minimum metric threshold such that at least 10 association rules are extracted from the frequent itemsets found (5 points)
- use "confidence" as metric and output the line below:
- Metric: "confidence" - min_metric: 0.xxxx - Number of rules: n

similarly as before, we will start from the maximum possible confidence, and decrease it until we found the required threshold

```python
min_rules_threshold = 10

min_confidence = .5

while min_confidence > 0:
    frequent_itemsets = apriori(basket, min_support=min_support, use_colnames=True)

    rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=min_confidence, num_itemsets=len(frequent_itemsets))

    if len(rules) >= min_rules_threshold:
        break

    min_confidence -= step

if min_confidence == 0:
    print(f"nothing found that matches the requirements, try with a bigger range")
else:
    print(f"found min_confidence for the given requirements")
    print(f"Metric: \"confidence\" - min_metric: {min_confidence:.4f} - number of rules: {len(rules)}")
```

### 7. print the first 10 rules found, sorted by descending confidence and support (3 points)

```python
sorted_rules = rules.sort_values(by=['confidence', 'support'], ascending=False).reset_index(drop=True)

sorted_rules.head(10)
```

### 8. plot confidence and support for all the sorted rules found (3 points)

```python
sorted_rules[['confidence', 'support']].plot(title="Association Rules")
```

### 9. scatter plot the rules by confidence and support, labelling the points with the index value of the corresponding rule
(hint https://stackoverflow.com/questions/14432557/matplotlibscatter-plot-with-different-text-at-each-data-point) (3 points)

```python
fig = sorted_rules.plot.scatter(x='confidence', y='support', title="association rules")

for i in range(len(sorted_rules)):
    fig.annotate(i, xy=(sorted_rules['confidence'][i], sorted_rules['support'][i]))
```
