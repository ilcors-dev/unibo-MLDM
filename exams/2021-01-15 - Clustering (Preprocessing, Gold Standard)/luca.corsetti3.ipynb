```python
# imports & setup

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, confusion_matrix, ConfusionMatrixDisplay

%matplotlib inline

random_state=777

np.random.seed(random_state)
```

### 1. load the data and separate in X all the columns but the last one, in y the last column, then produce a pairplot of X and comment what you see (4pt)

first of all let's read the dataset from the provided csv file, then we'll split the dataset into dataframes, one for the y and one for the X

```python
df = pd.read_csv('./exam_2021_01_15.csv')

y = df['y']

X = df.drop(columns='y')
```

```python
sns.pairplot(X)
```

given the pairplot we can clearly notice how column pairs (A0, A1) & (A1, A0) have 2 distinct groups of points. hinting that 2 clusters would be enough to categorize the data

### 2. find the best clustering scheme for X with a method of your choice, plot the silhouhette index for an appropriate range of parameters and show the chosen hyperparameter(s) (4pt)
    1. consider carefully the number of clusters, simple
    optimisation of the silhouette will not be enough, decide
    visually the best number of clusters

we will use KMeans to find the best clustering scheme for X, trying various range to find the optimal one

```python
silhouette_scores = []
k_range = range(2, 11)

for k in k_range:
    km = KMeans(n_clusters=k, random_state=random_state)

    y_km = km.fit_predict(X, y)
    score = silhouette_score(X, y_km)
    silhouette_scores.append(score)

    print(f"k={k}, silhouette_score={score:.2f}")
```

let's plot the various slihouette scores for the given clusters

```python
plt.plot(k_range, silhouette_scores, color='blue')
plt.xlabel('k clusters')
plt.ylabel('silhouette score', color='blue')
plt.tick_params(axis='y', labelcolor='blue')

plt.tight_layout()
plt.show()
```

the silhouette score is at its best at **k=2**, as hinted by looking at the initial pairplot.\
however, we notice that the raw **y** may assume values [0, 1, 2]. we will therefore choose **k=3** as the number of clusters.

```python
best_k = 3
```

### 3. fit the clustering scheme store the cluster labels in y_km and output the silhouette score (2pt)

```python
km = KMeans(n_clusters=best_k, random_state=random_state)

y_km = km.fit_predict(X)

print(f"the silhouette score for k={best_k} is {silhouette_score(X, y_km):.4f}")
```

### 4. use the labels in the last column of the input file as the “gold standard” for the clustering and compare y_km and y; for aneffective comparison, each label in y_km must be remapped to the best label in y; compute and apply this re-mapping (5pt)
    1. hint for each subset of the data with x in y_km find the
    most frequent label in y

not much clear. as of my understanding the question is asking us to re-map the predicted labels to the original ones provided in the y dataframe by looking at the most frequent ones. in other words, change the labels in y_km to the most frequent one found in the same range of y.\
what is a range? we consider a range to be a "streak" of the same labels of the y_km subset

```python
previous_label = y_km[0]

start_idx = 0
changes_count = 0

for i, label in enumerate(y_km):
    if label != previous_label:
        # count the occurrences in the range in the gold standard
        occurrences = np.bincount(y[start_idx:i])

        # find the most common y label in the range from the gold standard
        gold_standard = np.argmax(occurrences)

        for j in range(start_idx, i):
            if y_km[j] != gold_standard:
                y_km[j] = gold_standard

                changes_count += 1

        start_idx = i

    # after the processing, update the label to find a new range
    previous_label = label

print(f"this remapping performed {changes_count} changes to y_km subsets")
```

### 5. produce the confusion matrix comparing y and y_km with sklearn.metrics.confusion_matrix, (2pt)

```python
cm = confusion_matrix(y, y_km)

cm
```

```python
fig = ConfusionMatrixDisplay(cm)

fig.plot()
```

### 6. consider possible pre-processing actions, repeat the fitting and evaluate as before the result of the new fitting (8pt)

by plotting the confusion matrix, we can see that our clustering scheme tends to predict 0 and 1 a bit too much often\

we could try applying a logarithmic transformation to the columns that don't have values <= 0 and see if we obtain a different result

```python
for column in X.columns:
    # We don't want to transform columns with values
    # Lower than or equal to zero
    if len(X[column]) != sum(np.greater(X[column], 0)):
        continue
        
    X[column] = np.log(X[column])

X.head()
```
